```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols = ['fLength','fWidth','fSize','fConc','fConc1','fAsym','fM3Long','fM3Trans','fM3Alpha','fDist','class']
df = pd.read_csv('./magic04.data', names=cols)
df.head()
```

```python
df['class']=df['class'].apply(lambda e: 1 if e=='g' else 0)
df['class'].unique()
```

```python
# filtering the class -label (g-->1, h-->0)
df_1 = df[df['class']==1]
df_1

df_0 = df[df['class']==0]
df_0
```

```python
for label in cols:
    plt.hist(df_1[label], color='blue',  label='gamma',  alpha=0.6)
    plt.hist(df_0[label], color='green', label='hedron', alpha=0.6)

    plt.title(label)
    plt.xlabel(label)
    plt.ylabel('Probability')
    plt.show()
    
```
